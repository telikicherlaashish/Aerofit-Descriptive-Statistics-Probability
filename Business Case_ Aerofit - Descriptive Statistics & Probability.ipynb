```python
def fitness_types (val):
  if val == 1:
    return "Poor"
  elif val == 2:
    return "Okay"
  elif val == 3:
    return "Average"
  elif val == 4:
    return "Good"
  else:
    return "Excellent"

df['fitness_type'] = df['Fitness'].apply(fitness_types)

##Sample space = ("Poor", "OK", "Average", Good", "Excellent")
```

```python
plt.hist(df['fitness_type'], bins=5, edgecolor='black')
# Add labels and title
plt.xlabel( 'Fitness Type')
plt.ylabel ( 'Frequency')
plt.title('Distribution of Fitness Types')
plt.grid(True)
plt.tight_layout ()
# Display the histogram
plt.show()
```

```python
def usage_types (val):
  if val < 3 :
    return "Low Usage"
  elif val == 3:
    return "Moderate Usage"
  else:
    return "High Usage"

df ["Usage_type"] = df["Usage"].apply(usage_types)
##Sample space = ("Low Usage", "moderate Usage", "High Usage")
df[ 'Usage_type'].hist()
# Add labels and title
plt.xlabel( 'Usage')
plt.ylabel( 'No of user')
plt.title('Distribution on the basis of usage')
plt.grid(False)
plt.tight_layout()
# Display the histogram
plt.show()
```

### Computing Marginal & Conditional Probabilities:

Marginal Probability

```python
df['Product'].value_counts(normalize=True)
```

Conditional Probabilities



-Probability of each product given gender:

```python
def p_prod_given_gender(gender, print_marginal=False):

      if gender!= "Female" and gender != "Male":
        return "Invalid gender value."


      df1 = pd.crosstab(index=df['Gender'], columns=[df['Product']])
      p_781 = df1['KP781'][gender] / df1.loc[gender].sum()
      p_481 = df1['KP481'][gender] / df1.loc[gender].sum()
      p_281 = df1['KP281'][gender] / df1.loc[gender].sum()

      if print_marginal:
          print(f"P(Male): {df1.loc['Male'].sum()/len(df):.2f}")
          print(f"P(Female): {df1.loc['Female'].sum()/len(df):.2f}\n")

      print(f"P(KP781/{gender}): {p_781:.2f}")
      print(f"P(KP481/{gender}): {p_481:.2f}")
      print(f"P(KP281/{gender}): {p_281:.2f}\n")


p_prod_given_gender('Male', True)
p_prod_given_gender('Female')
```

Conditional Probabilities-


Probability of each product given gender

```python

def p_prod_given_gender(gender, print_marginal=False):
    # Check for valid gender input
    if gender not in ["Female", "Male"]:
        return "Invalid gender value."

    # Create a crosstab of Gender vs Product
    df1 = pd.crosstab(index=df['Gender'], columns=[df['Product']])

    # Initialize a dictionary to store probabilities
    probabilities = {}

    # Calculate the probabilities for each product
    for product in df1.columns:
        probabilities[product] = df1[product][gender] / df1.loc[gender].sum()

    # Print marginal probabilities if requested
    if print_marginal:
        print(f"P(Male): {df1.loc['Male'].sum() / len(df):.2f}")
        print(f"P(Female): {df1.loc['Female'].sum() / len(df):.2f}\n")
        for product, prob in probabilities.items():
            print(f"P({product}/{gender}): {prob:.2f}")

    return probabilities

# Example usage
print(p_prod_given_gender('Male', True))
print(p_prod_given_gender('Female'))
```

```python
def p_prod_given_mstatus(MaritalStatus, print_marginal=False):
    # Check for valid marital status input
    if MaritalStatus not in ["Single", "Partnered"]:
        return "Invalid marital status value."

    # Create a crosstab of MaritalStatus vs Product
    df1 = pd.crosstab(index=df['MaritalStatus'], columns=[df['Product']])

    # Initialize a dictionary to store probabilities
    probabilities = {}

    # Calculate the probabilities for each product
    for product in df1.columns:
        probabilities[product] = df1[product][MaritalStatus] / df1.loc[MaritalStatus].sum()

    # Print marginal probabilities if requested
    if print_marginal:
        print(f"P(Single): {df1.loc['Single'].sum() / len(df):.2f}")
        print(f"P(Partnered): {df1.loc['Partnered'].sum() / len(df):.2f}\n")
        for product, prob in probabilities.items():
            print(f"P({product}/{MaritalStatus}): {prob:.2f}")

    return probabilities

# Example usage
print(p_prod_given_mstatus('Single', True))
print(p_prod_given_mstatus('Partnered'))
```
